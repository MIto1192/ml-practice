**＜線形回帰モデル実装（ボストンデータセット用いて＞**

**１．ボストンデータセットの準備**

```python
#ボストンデータ使用の準備
from sklearn.datasets import load_boston
from pandas import  DataFrame
import numpy as np
    
```

```python
#ボストンデータをインスタンスにインポート
boston = load_boston()
```

```python
#データ情報表示
print(boston)
#変数の説明表示
print(boston['DESCR'])
```

```python
#カラム名つける
print(boston['feature_names'])
```

```python
#data変数の中身を確認
print(boston['data'])
```

```python
#target変数(目的変数)の中身を確認
print(boston['target'])
```

**２．データフレームの準備**

```python
#説明変数らをDataFrameに変換
df = DataFrame(data=boston.data, columns = boston.feature_names)
```

```python
#目的変数をDataFrameに変換
df['PRICE'] = np.array(boston.target)
```

```python
#最初の5行を表示
df.head(12)
```

**３．線形単回帰分析**

```python
#dfでカラムを指定してデータを表示
df[['RM']].head(12)
```

```python
#説明変数
#:は全行、RM列を抽出
#loc[]であること注意！！
#colum名指定は['NAME']注意！！
data = df.loc[:, ['RM']].values
```

```python
#dataリスト表示
data[0:5]
```

```python
#目的変数
target = df.loc[:, ['PRICE']].values
```

```python
#sklearnライブラリの線形回帰モデル使用宣言
from sklearn.linear_model import LinearRegression
```

```python
#オブジェクト生成
model = LinearRegression()
model.get_params()
```

```python
#fit関数でパラメータ推定
model.fit(data, target)
```

```python
#新たな予測
#[[x]]で囲む
model.predict([[6]])
```

重回帰分析実装
