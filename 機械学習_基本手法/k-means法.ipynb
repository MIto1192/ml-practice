# K-meansを用いたクラスタリング

```python
from sklearn.cluster import KMeans
import numpy as np

inits= np.array([[-3.0],
                 [0]])
data = (np.array([[-2.7,-1.3,0.7,3.5,5.1]])).reshape(-1,1)
km = KMeans(n_clusters=2,init=inits, n_init=30, max_iter=300, tol=1e-04, random_state=0)
clusters_sklearn = km.fit_predict(data)

print(km.cluster_centers_)
print(km.labels_)
```
